```python
import pandas
```

```python
import numpy as np
```

```python
import Recommenders as Recommenders
```

```python
triplets_file = '10000.txt'
```

```python
songs_metadata_file = 'song_data.csv'
```

```python
#Reading of data set and defining of data set into 3 columns i.e. user_id,song_id and listen_count
song_df_1 = pandas.read_table(triplets_file,header=None)
song_df_1.columns = ['user_id', 'song_id', 'listen_count']
```

```python
song_df_2 =  pandas.read_csv(songs_metadata_file)
```

```python
#To merge triplets_file and songs_metadata_file and drop the duplicate columns
song_df = pandas.merge(song_df_1, song_df_2.drop_duplicates(['song_id']), on="song_id", how="left")
```

```python
#Printing first 5 rows after merging the datasets
song_df.head()
```

```python
#Returns the the total length of this dataset.
len(song_df)
```

```python
#Here a subset of first 10,000 songs is used 
song_df = song_df.head(10000)
```

```python
#Then, we merge song and artist_name into one column
song_df['song'] = song_df['title'].map(str) + " - " + song_df['artist_name']
```

```python
#Aggregated by number of time a particular song is listened too in general by all users. 
song_grouped = song_df.groupby(['song']).agg({'listen_count': 'count'}).reset_index()
#To calculate the group_sum by summing the listen_count of each song. 
grouped_sum = song_grouped['listen_count'].sum()
```

```python
#This is to add a new column called percentage, and calculate this percentage by 
#dividing the listen_count by the sum of listen_count of all songs and then multiply by 100. 
song_grouped['percentage']  = song_grouped['listen_count'].div(grouped_sum)*100
song_grouped.sort_values(['listen_count', 'song'], ascending = [0,1])
```

```python
#the number of unique users and songs in our subset of data
users = song_df['user_id'].unique()
```

```python
len(users)
```

```python
songs = song_df['song'].unique()
len(songs)
```

```python
#Now, we have imported train_test_split function from sckit-learn library to divide dataset into training set and test set
from sklearn.model_selection import train_test_split
```

```python
train_data, test_data = train_test_split(song_df, test_size = 0.20, random_state=0)
print(train_data.head(5))
```

```python
#Based on the popularity of each song, create a recommender that accept a user_id as input 
#and output a list of recommended songs to that user
pm = Recommenders.popularity_recommender_py()
pm.create(train_data, 'user_id', 'song')
```

```python
user_id = users[5]
```

```python
pm.recommend(user_id)
```

```python
user_id = users[8]
```

```python
pm.recommend(user_id)
```

```python
#We have created an instance item similarity based recommender class and feed it with our training data.
is_model = Recommenders.item_similarity_recommender_py()
is_model.create(train_data, 'user_id', 'song')
```

```python
#Print the songs for the user in training data
user_id = users[5]
user_items = is_model.get_user_items(user_id)
#
print("------------------------------------------------------------------------------------")
print("Training data songs for the user userid: %s:" % user_id)
print("------------------------------------------------------------------------------------")

for user_item in user_items:
    print(user_item)

print("----------------------------------------------------------------------")
print("Recommendation process going on:")
print("----------------------------------------------------------------------")

#Recommend songs for the user using personalized model
is_model.recommend(user_id)
```
